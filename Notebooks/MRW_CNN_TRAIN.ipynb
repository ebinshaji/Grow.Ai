```python
import os
import shutil
import random

def split_data(input_dir, train_dir, test_dir, test_ratio=0.1):
    # Iterate through each folder in the input directory
    for folder_name in os.listdir(input_dir):
        folder_path = os.path.join(input_dir, folder_name)
        if os.path.isdir(folder_path):
            # Create corresponding train and test folders
            train_folder = os.path.join(train_dir, folder_name)
            test_folder = os.path.join(test_dir, folder_name)
            os.makedirs(train_folder, exist_ok=True)
            os.makedirs(test_folder, exist_ok=True)

            # Get list of files in the folder
            files = os.listdir(folder_path)
            num_test_files = int(len(files) * test_ratio)
            test_files = random.sample(files, num_test_files)

            # Move files to train or test folder based on the split
            for file in files:
                src_path = os.path.join(folder_path, file)
                if file in test_files:
                    dst_path = os.path.join(test_folder, file)
                else:
                    dst_path = os.path.join(train_folder, file)
                shutil.move(src_path, dst_path)

# Define input directory containing class folders
input_dir = "/content/Apple Scab"
# Define train and test directories
train_dir = "/content/apple/train"
test_dir = "/content/apple/test"

# Define the test ratio (10%)
test_ratio = 0.1

# Call the function to split the data
split_data(input_dir, train_dir, test_dir, test_ratio)

print("Data split completed!")
```

```python
import os
import torch
import torch.nn as nn
import torchvision.transforms as transforms
from torch.utils.data import DataLoader
from torchvision.datasets import ImageFolder
import matplotlib.pyplot as plt

# Define custom dataset class
class PlantDiseaseDataset(torch.utils.data.Dataset):
    def __init__(self, data_dir, transform=None):
        self.dataset = ImageFolder(data_dir, transform=transform)

    def __len__(self):
        return len(self.dataset)

    def __getitem__(self, idx):
        return self.dataset[idx]

# Define MRW CNN model (replace with your specific architecture)
class MRW_CNN(nn.Module):
    def __init__(self, num_classes):
        super(MRW_CNN, self).__init__()
        # Convolutional layers for feature extraction
        self.conv1 = nn.Conv2d(3, 16, kernel_size=3, stride=1, padding=1)  # Input channels=3 (RGB), output=16 filters, kernel size 3x3
        self.conv2 = nn.Conv2d(16, 32, kernel_size=3, stride=1, padding=1)  # 16 input channels, 32 filters
        self.conv3 = nn.Conv2d(32, 64, kernel_size=3, stride=1, padding=1)  # 32 input channels, 64 filters
        self.pool = nn.MaxPool2d(kernel_size=2, stride=2)  # Max pooling for dimensionality reduction
        self.relu = nn.ReLU()  # Activation function
        self.dropout = nn.Dropout(0.5)

        # Fully-connected layers for classification
        self.fc1 = nn.Linear(64 * 28 * 28, 128)  # Flatten output from previous layer, 128 hidden units
        self.fc2 = nn.Linear(128, num_classes)  # Output layer with num_classes neurons (disease classes)

    def forward(self, x):
        # Pass through convolutional layers with ReLU activation and pooling
        x = self.relu(self.conv1(x))
        x = self.pool(x)
        x = self.relu(self.conv2(x))
        x = self.pool(x)
        x = self.relu(self.conv3(x))
        x = self.pool(x)

        # Flatten and pass through fully-connected layers
        x = x.view(-1, 64 * 28 * 28)  # Flatten feature maps
        x = self.relu(self.fc1(x))
        x = self.fc2(x)

        return x

# Hyperparameters
learning_rate = 0.001
batch_size = 16
num_epochs = 8

# Data transformations (adjust as needed)
data_transforms = transforms.Compose([
    transforms.Resize((224, 224)),  # Resize to 224x224
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])  # ImageNet normalization
])

# Load training and validation datasets
train_data_dir = "/content/apple/train"
train_dataset = PlantDiseaseDataset(train_data_dir, transform=data_transforms)
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)

val_data_dir = "/content/apple/test"
val_dataset = PlantDiseaseDataset(val_data_dir, transform=data_transforms)
val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)

# Define model, loss function, and optimizer
model = MRW_CNN(num_classes=len(os.listdir(train_data_dir)))
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)

# Training loop
train_loss_history = []
val_loss_history = []
train_accuracy_history = []
val_accuracy_history = []

for epoch in range(num_epochs):
    # Train
    model.train()
    running_train_loss = 0.0
    correct_train_predictions = 0
    total_train_samples = 0
    for images, labels in train_loader:
        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_train_loss += loss.item()
        _, predicted = torch.max(outputs, 1)
        total_train_samples += labels.size(0)
        correct_train_predictions += (predicted == labels).sum().item()

    epoch_train_loss = running_train_loss / len(train_loader)
    epoch_train_accuracy = correct_train_predictions / total_train_samples
    train_loss_history.append(epoch_train_loss)
    train_accuracy_history.append(epoch_train_accuracy)

    # Validation
    model.eval()
    running_val_loss = 0.0
    correct_val_predictions = 0
    total_val_samples = 0
    with torch.no_grad():
        for images, labels in val_loader:
            outputs = model(images)
            loss = criterion(outputs, labels)

            running_val_loss += loss.item()
            _, predicted = torch.max(outputs, 1)
            total_val_samples += labels.size(0)
            correct_val_predictions += (predicted == labels).sum().item()

    epoch_val_loss = running_val_loss / len(val_loader)
    epoch_val_accuracy = correct_val_predictions / total_val_samples
    val_loss_history.append(epoch_val_loss)
    val_accuracy_history.append(epoch_val_accuracy)

    print(f"Epoch [{epoch+1}/{num_epochs}], Train Loss: {epoch_train_loss:.4f}, Train Accuracy: {epoch_train_accuracy:.4f}, Val Loss: {epoch_val_loss:.4f}, Val Accuracy: {epoch_val_accuracy:.4f}")

# Plot the loss and accuracy of the model during training
plt.plot(train_loss_history, label='Training Loss')
plt.plot(val_loss_history, label='Validation Loss')
plt.plot(train_accuracy_history, label='Training Accuracy')
plt.plot(val_accuracy_history, label='Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Loss/Accuracy')
plt.legend()
plt.show()

# Save trained model
torch.save(model.state_dict(), "plant1.pth")
```

```python
import os
import torch
import torch.nn as nn
import torchvision.transforms as transforms
from torch.utils.data import DataLoader
from torchvision.datasets import ImageFolder
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

# Define custom dataset class
class PlantDiseaseDataset(torch.utils.data.Dataset):
    def __init__(self, data_dir, transform=None):
        self.dataset = ImageFolder(data_dir, transform=transform)
        self.classes = len(self.dataset.classes)  # Get the number of classes in the dataset

    def __len__(self):
        return len(self.dataset)

    def __getitem__(self, idx):
        return self.dataset[idx]

# Define MRW CNN model with Dropout regularization
class MRW_CNN(nn.Module):
    def __init__(self, num_classes):
        super(MRW_CNN, self).__init__()
        self.conv1 = nn.Conv2d(3, 16, kernel_size=3, stride=1, padding=1)
        self.conv2 = nn.Conv2d(16, 32, kernel_size=3, stride=1, padding=1)
        self.conv3 = nn.Conv2d(32, 64, kernel_size=3, stride=1, padding=1)
        self.pool = nn.MaxPool2d(kernel_size=2, stride=2)
        self.relu = nn.ReLU()
        self.dropout = nn.Dropout(0.5)  # Dropout layer with 50% probability

        self.fc1 = nn.Linear(64 * 28 * 28, 128)
        self.fc2 = nn.Linear(128, num_classes)

    def forward(self, x):
        x = self.relu(self.conv1(x))
        x = self.pool(x)
        x = self.relu(self.conv2(x))
        x = self.pool(x)
        x = self.relu(self.conv3(x))
        x = self.pool(x)

        x = x.view(-1, 64 * 28 * 28)
        x = self.relu(self.fc1(x))
        x = self.dropout(x)  # Apply dropout before the output layer
        x = self.fc2(x)

        return x

# Hyperparameters
learning_rate = 0.001
batch_size = 16
num_epochs = 30

# Data transformations (including data augmentation)
train_transforms = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.RandomRotation(degrees=30),
    transforms.RandomHorizontalFlip(),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])

val_transforms = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])

# Load dataset and split into train/validation sets
data_dir = "/content/apple"
train_dataset = PlantDiseaseDataset(os.path.join(data_dir, "train"), transform=train_transforms)
val_dataset = PlantDiseaseDataset(os.path.join(data_dir, "test"), transform=val_transforms)

train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)

# Define model, loss function, and optimizer
model = MRW_CNN(num_classes=train_dataset.classes)  # Pass the number of classes directly to the model
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)

# Training loop
train_loss_history = []
val_loss_history = []
train_accuracy_history = []
val_accuracy_history = []

for epoch in range(num_epochs):
    # Train
    model.train()
    running_train_loss = 0.0
    correct_train_predictions = 0
    total_train_samples = 0
    for images, labels in train_loader:
        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_train_loss += loss.item()
        _, predicted = torch.max(outputs, 1)
        total_train_samples += labels.size(0)
        correct_train_predictions += (predicted == labels).sum().item()

    epoch_train_loss = running_train_loss / len(train_loader)
    epoch_train_accuracy = correct_train_predictions / total_train_samples
    train_loss_history.append(epoch_train_loss)
    train_accuracy_history.append(epoch_train_accuracy)

    # Validation
    model.eval()
    running_val_loss = 0.0
    correct_val_predictions = 0
    total_val_samples = 0
    with torch.no_grad():
        for images, labels in val_loader:
            outputs = model(images)
            loss = criterion(outputs, labels)

            running_val_loss += loss.item()
            _, predicted = torch.max(outputs, 1)
            total_val_samples += labels.size(0)
            correct_val_predictions += (predicted == labels).sum().item()

    epoch_val_loss = running_val_loss / len(val_loader)
    epoch_val_accuracy = correct_val_predictions / total_val_samples
    val_loss_history.append(epoch_val_loss)
    val_accuracy_history.append(epoch_val_accuracy)

    print(f"Epoch [{epoch+1}/{num_epochs}], Train Loss: {epoch_train_loss:.4f}, Train Accuracy: {epoch_train_accuracy:.4f}, Val Loss: {epoch_val_loss:.4f}, Val Accuracy: {epoch_val_accuracy:.4f}")

# Plot the loss and accuracy of the model during training
plt.plot(train_loss_history, label='Training Loss')
plt.plot(val_loss_history, label='Validation Loss')
plt.plot(train_accuracy_history, label='Training Accuracy')
plt.plot(val_accuracy_history, label='Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Loss/Accuracy')
plt.legend()
plt.show()

# Save trained model
torch.save(model.state_dict(), "plant1.pth")
```
